```python
from ipywidgets import interactive

import m8r
spike = m8r.spike(n1=5,n2=5,n3=5,nsp=5,k1=[1,2,3,4,5],k3=[1,2,3,4,5])[0]
```

```python
def plot_slice(f3):
    slic = m8r.window(n3=1,f3=f3)[spike]
    return m8r.grey(title="Slice %d" % f3)[slic]
    
interactive_plot = interactive(plot_slice, f3=(0,spike.int('n3')-1))
```

```python
interactive_plot
```

```python
byte = m8r.byte(gainpanel='all')[spike]

def grey3(frame1,frame2,frame3):
    return m8r.grey3(title="Cube",flat=False,
                     frame1=frame1,frame2=frame2,frame3=frame3)[byte]
    
interactive_plot = interactive(grey3, 
                               frame1=(0,spike.int('n1')-1),
                               frame2=(0,spike.int('n2')-1),
                               frame3=(0,spike.int('n3')-1))
```
